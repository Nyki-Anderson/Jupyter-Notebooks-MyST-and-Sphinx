# Introduction to Lua

- Lua is a lightweight, high-level scripting language that was developed in the early 1990s.
- Lua is known for its simplicity, flexibility, and extensibility, making it a popular choice for various applications.

Lua is used in various domains and applications, including: 

- Game Development: Lua is widely used in the gaming industry as a scripting language for game engines and frameworks. Popular games like World of Warcraft and Angry Birds have employed Lua for their scripting needs.
- Embedded Systems: Lua' lightweight nature makes it suitable for embedded systems and IoT devices.
- Scripting and Automation: Lua is often used as a scripting language for automation tasks, providing an easy-to-use interface.
- Web Development: Lua has frameworks that allow developers to create high-performance web applications by embedding Lua within web servers.

# Interlude - Factorial Computation

On this lecture we are going to implement a function called _factorial_ that given a natural number _n_, it return the factorial of this number, it returns the factorial of this number.

The factorial _n_ is the product of all integers less than or equal to _n_.



## Recursive Function

A **recursive function** is a function that calls itself from within its own definition. It is a way of solving a problem by breaking it down into smaller, similar sub-problems and solving them recursively until a base case is reached.

1. **Base case:** Condition that specifies when the recursion should stop.
2. **Recursive case:** Part where the function calls itself with a modified input, moving closer to the base case.

````lua
%save -rf factorial.lua
-- Recursive Factorial
function factorial(n)
    if n == 0 then -- Base Case
        return 1
    else 
        return n * factorial(n-1)
    end
end
````

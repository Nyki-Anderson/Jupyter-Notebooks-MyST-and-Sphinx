# Building a Custom Jupyter Notebook Kernel

I found when starting the [Lua Programming Udemy course](https://www.udemy.com/share/10a9Qq3@bsv9B0_-y2oIqI8dNeFs0eOv7PG5VL1JESM-yN0Mw9HtaVPeFRODH5bI3g_TOZw=/) that the stock `xeus-lua` kernel was not dynamic enough to allow so-called _magic_ commands offered by the IPython Kernel. Specifically, I could not run the `lua` code as files that receive user input. 

In reaction --and at risk of becoming too granular-- I am taking this opportunity to learn how to build my own custom Jupyter kernel. 

:::{important} Objectives
The objective of this exercise is to build a custom `lua` kernel that allows for the use of magic `shell` commands within the `lua` code cells.
:::

## [Using `xeus` to Author Custom Kernels](https://xeus.readthedocs.io/en/latest/usage.html)

`xeus` enables custom kernel authors to implement Jupyter kernels more easily. It takes the burden of implementing the Jupyter Kernel protocol so developers can focus on implementing the interpreter part of the kernel.
